# Roll A — Data Loading

```python
import pandas as pd
from supabase import create_client
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")

print("URL exists:", SUPABASE_URL is not None)
print("KEY exists:", SUPABASE_KEY is not None)
```

```python
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

print("Supabase connected")
```

```python

# Kqigi andmete laadimine sales tabelist
all_sales = []
offset = 0
while True:
    response = supabase.table('sales').select('*').range(offset, offset + 999).execute()
    if not response.data:
        break
    all_sales.extend(response.data)
    offset += 1000

df_sales = pd.DataFrame(all_sales)
print("Sales shape:", df_sales.shape)
df_sales.head()
```

```python
# Kqigi andmete laadimine customers tabelist
all_customers = []
offset = 0
while True:
    response = supabase.table('customers').select('*').range(offset, offset + 999).execute()
    if not response.data:
        break
    all_customers.extend(response.data)
    offset += 1000

df_customers = pd.DataFrame(all_customers)
print("Customers shape:", df_customers.shape)
df_customers.head()
```

```python
df = pd.merge(
    df_sales,
    df_customers,
    on="customer_id",
    how="left"
)

print("Merged shape:", df.shape)

df.head()
```

```python
print("Merged DataFrame shape:")
print(df.shape)

print("\nData types:")
print(df.dtypes)

print("\nFirst 5 rows:")
display(df.head())
```

```python
required_columns = ["customer_id", "sale_date", "total_price", "email"]

for column in required_columns:
    print(column, "exists:", column in df.columns)
```

# ROLL B - Data Cleaning

```python
print("Initial shape:", df.shape)
```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
print("NULL values:\n", df.isnull().sum())
```

```python
df = df.dropna(subset=['customer_id', 'sale_date', 'total_price'])
print("After removing critical NULLs:", df.shape)
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
print("sale_date type:", df['sale_date'].dtype)
```

```python
print("Negative total_price values:", (df['total_price'] <= 0).sum())
df = df[df['total_price'] > 0]
print("After removing outliers:", df.shape)
```

```python
print("--- CLEANING REPORT ---")
print(f"Final shape: {df.shape}")
print(f"Unique customers: {df['customer_id'].nunique()}")
print(f"Date range: {df['sale_date'].min().date()} to {df['sale_date'].max().date()}")
```
